```python
import pandas as pd #import pandas library
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0] # read wikipedia page as df
```

```python
df.head()
```

```python
df.columns = df.iloc[0] #Remove 0, 1, 2 from column header
df = df.reindex(df.index.drop(0)).reset_index(drop=True)
df.columns.name = None
df.columns = ['PostalCode', 'Borough', 'Neighbourhood'] # rename column postcode as PostalCode
df.head()
```

```python
indexNames = df[ df['Borough'] == 'Not assigned' ].index # drop rows with Boroughs = not assigned
df.drop(indexNames, inplace = True)
df.shape
```

```python
df
```

```python
post = df['PostalCode'].unique().tolist() # count number of unique postal codes to a list
```

```python
len(post) #determine length of list post
```

```python
df.replace({'Neighbourhood': 'Not assigned'}, df['Borough'])
df.head()
df.shape
```

```python
# merge neighbourhood with same Postalcode
grouped_df = df.groupby(['PostalCode', 'Borough'])['Neighbourhood'].apply(lambda Neighbourhood: ','.join(Neighbourhood))
grouped_df = grouped_df.reset_index()
grouped_df
```

```python
new_df = df.merge(grouped_df, how = 'right')
new_df
```

```python
new_df.shape
```

```python
lat_long_df = pd.read_csv('https://cocl.us/Geospatial_data')# read csv file for latitude, longitude
#rename column Postal Code as PostalCode so that it is the same as original df
lat_long_df.columns= ['PostalCode','Latitude','Longitude']
lat_long_df.head()
```

```python
combined_df = new_df.merge(lat_long_df, on = 'PostalCode', how = 'inner')#merge the 2 dataframes
combined_df
```
